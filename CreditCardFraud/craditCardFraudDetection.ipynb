```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("creditcard.csv")
```

```python
df.shape
```

```python
df.iloc[:,-1].value_counts()
```

```python
def transform(x):
    return (x-x.min())/(x.max()-x.min())
```

```python
df['Time']=transform(df['Time'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import RobustScaler
rb=RobustScaler()

df['Amount']=rb.fit_transform(df['Amount'].to_numpy().reshape(-1,1))
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=.2,random_state=2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(max_depth=5)
```

```python
dt.fit(X_train,y_train)
```

```python
y_prd=dt.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,f1_score
```

```python
confusion_matrix(y_test,y_prd)
```

```python
from sklearn.svm import SVC
```

```python
sv=SVC()
```

```python
sv.fit(X_train,y_train)
```

```python
y_pred=sv.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import precision_score,recall_score
```

```python
print(f"sv_pre={precision_score(y_test,y_pred,average="macro")}")
print(f"sv_rec={recall_score(y_test,y_pred,average="macro")}")
print(f"dt_pre={precision_score(y_test,y_prd,average="macro")}")
print(f"dt_rec={recall_score(y_test,y_prd,average="macro")}")
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression 
```

```python
estimators=[
    ('rf',RandomForestClassifier(n_estimators=50)),
    ('gd',GradientBoostingClassifier()),
    ('svc',SVC()),
    
]
```

```python
stclf=StackingClassifier(estimators=estimators, final_estimator=LogisticRegression(), cv=5, n_jobs=-1,)
```

```python
stclf.fit(X_train,y_train)
```

```python
y_pred=stclf.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(f"stclf_pre={precision_score(y_test,y_pred,average="macro")}")
print(f"stcl_rec={recall_score(y_test,y_pred,average="macro")}")
print(f"f1_sc={f1_score(y_test,y_pred)}")
```

# So now we get the best f1_sc, precision and recall value so we select StackingClassifier
