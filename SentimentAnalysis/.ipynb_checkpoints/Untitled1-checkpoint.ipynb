```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
sa=SentimentIntensityAnalyzer()
```

```python
sa.polarity_scores("This product is very good")
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('movies.csv')
```

```python
df.head(5)
```

```python
df=df.dropna()
```

```python
df.insert(0,"Id",[i for i in range(df.shape[0])],True)
```

```python
res={}
for i, row in tqdm(df.iterrows(),total=len(df)):
    review=row['reviewText']
    myId=row['Id']
    try:
        score=sa.polarity_scores(review)
        res[myId]=score
    except(AttributeError):
        print(f"error with id {myId}")
```

```python
res
```

```python
res=pd.DataFrame(res).T
```

```python
res.insert(0,"Id",[i for i in range(res.shape[0])],True)
```

```python
import random
reviews=[]
for i in range(len(res)):
    if(res.iloc[i,2]<7.5 and res.iloc[i,3]<.15 ):
        r = random.randint(1,2)
        if(r==1):
            reviews.append('POSITIVE')
        else:
            reviews.append('NEGATIVE')
    else:
        if(res.iloc[i,1]<res.iloc[i,3]):
            reviews.append('POSITIVE')
        else:
            reviews.append('NEGATIVE')
```

```python
reviews
```

```python
import copy
df1=copy.copy(df)
```

```python
df1.insert(6,'P_reviews',reviews,True)
```
