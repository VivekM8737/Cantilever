```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('reviews.csv')
```

```python
df.head()
```

```python
df['feedback']=df['Review']+" "+df['Summary']
```

```python
df=df.dropna()
```

```python
import string
```

```python
punc=string.punctuation
```

```python
punc
```

```python
def remove_(text):
    return text.translate(str.maketrans('','',punc))
```

```python
df['feedback']=df['feedback'].apply(remove_)
```

```python
df.head()
```

```python
def lc(text):
    return text.lower()
```

```python
df['feedback']=df['feedback'].apply(lc)
```

```python
from nltk.corpus import stopwords
```

```python
stop_w=stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stop_w(text):
    new_l=[]
    for word in text.split():
        if word not in stop_w.words('english'):
            new_l.append(word)
    x=new_l[:]
    new_l.clear()
    return " ".join(x)
    
```

```python
df['feedback']=df['feedback'].apply(remove_stop_w)
```

```python
df.head()
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
df['feedback']=df['feedback'].apply(stem_words)
```
